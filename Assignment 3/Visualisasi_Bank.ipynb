```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Bank.csv", sep=';')
to_drop = ['day', 
           'month',
           'pdays',
           'previous',
           'default',
           'balance',
           'housing',
           'loan',
           'contact',
           'duration',
           'campaign',
           'poutcome',
           'age'
           ]
data.drop(to_drop, inplace=True, axis=1)
data.head()
```

```python
data.info()
```

```python
unique_job = data[data.y=='yes']['job'].unique()
unique_job.sort()
unique_job
```

```python
job_subscription = []
for i in unique_job:
  job_subscription.append(data[data.y=='yes'][data['job']==i].y.count())
```

```python
print(set(zip(unique_job, job_subscription)))
```

```python
unique_marital = data[data.y=='yes']['marital'].unique()
unique_marital
```

```python
marital_subscription = []
for i in unique_marital:
  marital_subscription.append(data[data.y=='yes'][data['marital']==i].y.count())
```

```python
print(set(zip(unique_marital, marital_subscription)))
```

```python
unique_education = data[data.y=='yes']['education'].unique()
unique_education
```

```python
education_subscription = []
for i in unique_education:
  education_subscription.append(data[data.y=='yes'][data['education']==i].y.count())
```

```python
print(set(zip(unique_education, education_subscription)))
```

```python
plt.barh(unique_job, job_subscription)
plt.show()
```

```python
plt.barh(unique_education, education_subscription)
plt.show()
```

```python
plt.barh(unique_marital, marital_subscription)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(job_subscription)
plt.legend(unique_job, loc='best')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(marital_subscription)
plt.legend(unique_marital, loc='best')
plt.show()
```
