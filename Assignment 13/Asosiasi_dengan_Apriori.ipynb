```python
# Langkah 1: Mengimpor library yang diperlukan
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Langkah 2: Meload dan Mengeksplor data
# Memuat Data
data = pd.read_csv("data_transaksi_restoran.csv")
data.head(10)
```

```python
# Mengeksplor kolom data
data.columns
```

```python

'''
data['Description'] = data['Description'].str.strip()

# Menghapus baris tanpa nomor faktur (InvoiceNo)
data.dropna(axis = 0, subset =['InvoiceNo'], inplace = True)
data['InvoiceNo'] = data['InvoiceNo'].astype('str')

# Menghapus semua transaksi yang dilakukan secara kredit
data = data[~data['InvoiceNo'].str.contains('C')]
'''

data['TRX_TS'] = data['TRX_TS'].str.extract(r'(.{8})').astype('str')
data['TRX_TS'] = data['TRX_TS'].str.strip()
data['TRX_TS'] = data['TRX_TS'].str.replace('\\','/')
print(data['TRX_TS'])
data.head(10)
```

```python
data.TRX_TS.unique()
```

```python
# Transaksi Tanggal 05/31/17
TS_05_31_17 = (data[data['TRX_TS'] =='05/31/17']
          .groupby(['TRX_ID', 'PRODUCT_NAME'])['SALES']
          .sum().unstack().reset_index().fillna(0)
          .set_index('TRX_ID'))
TS_05_31_17
```

```python
# Transaksi Tanggal 06/02/20
TS_06_02_17 = (data[data['TRX_TS'] =='06/02/20']
          .groupby(['TRX_ID', 'PRODUCT_NAME'])['SALES']
          .sum().unstack().reset_index().fillna(0)
          .set_index('TRX_ID'))
TS_06_02_17
```

```python
# Transaksi Tanggal 06/03/20
TS_06_03_20 = (data[data['TRX_TS'] =='06/03/20']
          .groupby(['TRX_ID', 'PRODUCT_NAME'])['SALES']
          .sum().unstack().reset_index().fillna(0)
          .set_index('TRX_ID'))
TS_06_03_20
```

```python
def hot_encode(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1
```

```python
trs_encoded = TS_05_31_17.applymap(hot_encode)
TS_05_31_17 = trs_encoded

trs_encoded = TS_06_02_17.applymap(hot_encode)
TS_06_02_17 = trs_encoded

trs_encoded = TS_06_03_20.applymap(hot_encode)
TS_06_03_20 = trs_encoded
TS_06_03_20
```

```python
frq_items = apriori(TS_05_31_17, min_support = 0.05, use_colnames = True)

# Mengumpulkan aturan yang disimpulkan dalam dataframe
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
#print(rules.head())
rules.head()
```

```python
frq_items = apriori(TS_06_02_17, min_support = 0.01, use_colnames = True)
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
#print(rules.head())
rules.head(10)
```

```python
frq_items = apriori(TS_06_03_20, min_support = 0.05, use_colnames = True)
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
#print(rules.head())
rules.head(6)
```
